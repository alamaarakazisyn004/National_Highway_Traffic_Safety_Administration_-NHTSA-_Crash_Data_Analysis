Import Required Libraries

```python
import pandas as pd
import numpy as np
```

Load Dataset

```python
df = pd.read_csv("/content/raw_crash_data.csv")

df.head()
```

```python
df.info()
df.describe()
df.shape
```

Data Cleaning

```python
# Remove Unnecessary Columns
df.drop(columns=[
'unique_id',
'case_id_fkey',
'publish',
'point',
'data_loaded_at'
], inplace=True)
```

```python
# Convert Date Columns
df['collision_datetime'] = pd.to_datetime(df['collision_datetime'])
df['death_datetime'] = pd.to_datetime(df['death_datetime'])
```

Handle Missing Values

```python
# Remove Rows with Missing Values
df = df.dropna(subset=[
'death_time',
'collision_time',
'age',
'analysis_neighborhood',
'supervisor_district',
'data_as_of'
])
```

```python
df.isnull().sum()
```

Extract Time Features

```python
df['hour'] = pd.to_datetime(df['collision_time']).dt.hour
df['day'] = df['collision_datetime'].dt.day
df['month'] = df['collision_datetime'].dt.month
df['year'] = df['collision_datetime'].dt.year
```

Create Severity Target

```python
df['severity'] = df['deceased'].apply(lambda x: 1 if x == True else 0)
```

Save Cleaned Dataset

```python
df.to_csv("/content/cleaned_crash_data.csv", index=False)
```
