Import Required Libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set_style("whitegrid")

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

Load Dataset

```python
df = pd.read_csv("/content/cleaned_crash_data.csv")

df.head()
```

Exploratory Data Analysis (EDA)


```python
# 1. Accidents by Year

sns.countplot(x='collision_year', data=df)
plt.title("Accidents by Year")
plt.show()
```

```python
# 2. Accidents by Gender

sns.countplot(x='sex', data=df)
plt.title("Accidents by Gender")
plt.show()
```

```python
# 3. Age Distribution

sns.histplot(df['age'], bins=30)
plt.title("Age Distribution in Accidents")
plt.show()
```

```python
# 4. Accidents by Collision Type

sns.countplot(y='collision_type', data=df)
plt.title("Collision Type Distribution")
plt.show()
```

```python
# 5. Accidents by Hour of Day

sns.countplot(x='hour', data=df)
plt.title("Accidents by Hour")
plt.show()
```

```python
# 6. Neighborhood Accident Distribution

df['analysis_neighborhood'].value_counts().plot(kind='bar')
plt.title("Accidents by Neighborhood")
plt.show()
```
